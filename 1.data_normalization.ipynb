* X_train: input data with 3 time series and 5 time steps

```
X_train = np.array([[ 1., -1.,  2., 3,  1],
...                     [ 2.,  0.,  0., 1, 2],
...                     [ 0.,  1., -1., -1, 3]])

X_test = np.array([[ 1., 0.,  1., -1,  1],
...                     [ 2.,  1.,  3., 1, 2],
...                     [ 1.,  1., -1., 1, -2]])
```

* **Standarlization** data with zero mean and unit variance.

```
scaler = preprocessing.StandardScaler().fit(X_train.T)
scaler
```

```
# mean
print(scaler.mean_)
# std
print(scaler.scale_)
```

```
# standarlized training data
scaler.transform(X_train.T).T
```

```
# standarlized test data
scaler.transform(X_test.T).T
```
